```python
import pandas as pd
import re
```

```python
train_set = pd.read_csv("./jigsaw_train_set.csv")
```

```python
train_set
```

```python
# Rimozione "\r" e "\n"
train_set["comment_text"] = train_set["comment_text"].apply(lambda x: re.sub(r'[\r\n]+', '', x))
```

```python
# Rimozione "::+"
train_set["comment_text"] = train_set["comment_text"].apply(lambda x: re.sub(r'::+', '', x))
```

```python
# Rimozione "==+"
train_set["comment_text"] = train_set["comment_text"].apply(lambda x: re.sub(r'==+', '', x))
```

```python
# Rimozione "**+"
train_set["comment_text"] = train_set["comment_text"].apply(lambda x: re.sub(r'\*\*+', '', x))
```

```python
# Rimozione IP Address
train_set["comment_text"] = train_set["comment_text"].apply(lambda x: re.sub(r'\b(?:\d{1,3}\.){3}\d{1,3}\b', '', x))
```

```python
# Rimozione Contenuto tra Parentesi Quadre
train_set["comment_text"] = train_set["comment_text"].apply(lambda x: re.sub(r'\[\[[^\[\]]+\]\]', '', x))
```

```python
# Rimozione Apici
train_set["comment_text"] = train_set["comment_text"].apply(lambda x: re.sub(r"['\"]", "", x))
```

```python
# Splitting di token in cui compare un segno di interpunzione forte seguito da una lettera maiuscola
train_set["comment_text"] = train_set["comment_text"].apply(lambda x: re.sub(r'([?!\.])([A-Z]\w*)', r'\1 \2', x))
```

```python
# Splitting di "combined words" (parolaParola)
train_set["comment_text"] = train_set["comment_text"].apply(lambda x: re.sub(r'([a-z])([A-Z])', r'\1 \2', x))
```

```python
train_set
```

```python
train_set.to_csv("./train_data.csv", index=False)
```

```python
test_set = pd.read_csv("./jigsaw_test_set.csv")
```

```python
test_set["comment_text"] = test_set["comment_text"].apply(lambda x: re.sub(r'[\r\n]+', '', x))
test_set["comment_text"] = test_set["comment_text"].apply(lambda x: re.sub(r'::+', '', x))
test_set["comment_text"] = test_set["comment_text"].apply(lambda x: re.sub(r'==+', '', x))
test_set["comment_text"] = test_set["comment_text"].apply(lambda x: re.sub(r'\*\*+', '', x))
test_set["comment_text"] = test_set["comment_text"].apply(lambda x: re.sub(r'\b(?:\d{1,3}\.){3}\d{1,3}\b', '', x))
test_set["comment_text"] = test_set["comment_text"].apply(lambda x: re.sub(r'\[\[[^\[\]]+\]\]', '', x))
test_set["comment_text"] = test_set["comment_text"].apply(lambda x: re.sub(r"['\"]", "", x))
test_set["comment_text"] = test_set["comment_text"].apply(lambda x: re.sub(r'([?!\.])([A-Z]\w*)', r'\1 \2', x))
test_set["comment_text"] = test_set["comment_text"].apply(lambda x: re.sub(r'([a-z])([A-Z])', r'\1 \2', x))
```

```python
test_set.to_csv("./test_data.csv", index=False)
```
