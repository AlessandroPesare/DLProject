# GUARDA DOPO

```python
# Lemmatizzazione
nltk.download('wordnet')
nltk.download('punkt')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lista = training_set["comment_text"].to_list()
```

```python
lista
```

```python
comment_text = list()
```

```python
from nltk.tag import pos_tag
nltk.download('averaged_perceptron_tagger')
```

```python
for frase in lista:
    tokens = word_tokenize(frase)
    lemmatized_tokens = list()
    for token in tokens:
        pos = pos_tag([token])[0][1][0].lower() if wordnet.synsets(token) else 'n'
        try:
            lemmatized_token = lemmatizer.lemmatize(token, pos=pos)
            lemmatized_tokens.append(lemmatized_token)
        except KeyError:
            # Se il POS non è valido, usa 'n' come default per la lemmatizzazione
            lemmatized_tokens.append(token)

    new_frase = ' '.join(lemmatized_tokens)
    comment_text.append(new_frase)
```

```python
training_set2 = training_set.copy()
```

```python
training_set2['comment_text'] = comment_text
```

# VETTORIZZAZZIONE

```python
# Creazione Vettori TFIDF
vectorizer = TfidfVectorizer()
```

```python
comment_text = training_set['comment_text']
type(comment_text)
```

```python
X_train = vectorizer.fit_transform(comment_text)
```

```python
y_train = training_set['toxic']
```

```python
y_train
```

```python
rf_classifier = RandomForestClassifier()
```

```python
rf_classifier.fit(X=X_train, y=y_train)
```

```python
rf_classifier.n_classes_
```

```python
y_pred = rf_classifier.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_true=y_train, y_pred=y_pred))
```
